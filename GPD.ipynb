```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = '/content/GDP.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Sorting the data by year to ensure the bubbles are ordered from 2016 to 2023
df_sorted = df.sort_values(by='Year')

# Data for plotting
x_sorted = df_sorted['Economic']
y_sorted = df_sorted['GDP (CVM)']
size_sorted = abs(df_sorted['GDP (CVM)']) * 100  # Use absolute value for size
years_sorted = df_sorted['Year']

# Assigning colors based on positive and negative GDP values
colors = ['green' if gdp > 0 else 'red' for gdp in df_sorted['GDP (CVM)']]

# Plotting the bubble chart with color-coding for positive and negative GDP
plt.figure(figsize=(10, 6))
plt.scatter(x_sorted, y_sorted, s=size_sorted, alpha=0.5, edgecolor='k', marker='o', c=colors)

# Adding labels for each year
for i, year in enumerate(years_sorted):
    plt.text(x_sorted.iloc[i], y_sorted.iloc[i], str(year), fontsize=9, ha='center', va='center')

# Adding titles and labels
plt.title('Relation between Budget of Economic and GDP (2016-2023)', fontsize=16)
plt.xlabel('Budget of Economic (Million Bath)', fontsize=12)
plt.ylabel('GDP', fontsize=12)

# Displaying the plot
plt.show()
```
